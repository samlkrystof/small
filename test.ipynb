```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

def visualize_initialization(tensor, title):
    plt.figure(figsize=(10, 5))
    plt.hist(tensor.detach().numpy().flatten(), bins=50)
    plt.title(title)
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    
# Create sample layers
embedding = nn.Embedding(1000, 64)
linear = nn.Linear(64, 64)

# Visualize initializations
print(f"Embedding weight stats:")
print(f"Mean: {embedding.weight.mean().item():.4f}")
print(f"Std: {embedding.weight.std().item():.4f}")
print(f"\nLinear weight stats:")
print(f"Mean: {linear.weight.mean().item():.4f}")
print(f"Std: {linear.weight.std().item():.4f}")
```

```python
visualize_initialization(embedding.weight, 'Embedding Weight Initialization')
```

```python
visualize_initialization(linear.weight, 'Linear Weight Initialization')
```
